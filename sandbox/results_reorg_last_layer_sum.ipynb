```python
def last_layer_df2latex(metric):

    """a func for converting the summary df to latex"""
    
    # Read the CSV data (replace 'csv_file.csv' with your actual file path)
    df = pd.read_csv(
        f"results/summary/all_results_last_layer/{metric}/last_layer_summary.csv"
    )

    # Drop the index column if it's the first column
    df = df.drop(columns=[df.columns[0], "model_layer"])

    # Replace 'True' with blank and 'False' with 'F'
    df.replace({True: "", False: "F"}, inplace=True)

    # Format 'best_layer_percent' column to two decimal places
    df["best_layer_percent"] = df["best_layer_percent"].apply(lambda x: f"{x:.2f}")
    # reorder column
    df = df[
        [
            "task",
            "model",
            "Emb > OH",
            "Emb > RI",
            "Emb > ST",
            "RI > OH",
            "ST > OH",
            "RI > ST",
            "best_layer",
            "best_layer_percent",
        ]
    ]

    df = df.rename(
        columns={
            "task": "Task",
            "model": "Model",
            "best_layer": "Best layer",
            "best_layer_percent": "Best layer percent",
        }
    )

    # Convert to LaTeX without including the index
    return df.to_latex(index=False)
        
```
